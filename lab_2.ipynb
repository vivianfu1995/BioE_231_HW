```python
# store all sequence names in their corresponding cluster

cluster_1 = ['hu.39','rh.50','rh.57','rh.51','rh.49','rh.53','rh.64','rh.61','rh.58']
cluster_2 = ['pi.1','pi.2','pi.3']
cluster_3 = ['rh.10','bb.1','bb.2','hu.17','hu.6']
cluster_4 = ['rh.2']
cluster_5 = ['hu.67','hu.40','hu.37','hu.42','hu.41','hu.66','rh.38','rh.40']
cluster_6 = ['rh.13','cy.5','cy.6','cy.4','cy.3']
cluster_7 = ['rh.35','rh.36','rh.37']
cluster_8 = ['rh.54','rh.55','rh.48','rh.62','cy.2']
cluster_9 = ['hu.48','hu.44','hu.46','hu.43']
cluster_10 = ['hu.14','hu.31','hu.32','hu.14']
cluster_11 = ['rh.43']
cluster_12 = ['rh.52']

clusters = [cluster_1,cluster_2,cluster_3,cluster_4,cluster_5,cluster_6,cluster_7,cluster_8,cluster_9,cluster_10,cluster_11,cluster_12]
```

```python
# create a list of dictionaries called clusters_seq_dict: 
# each dictionary is a cluster with keys being the sequence id and values being the sequence

clusters_seq_dict = []
for cluster in clusters:
    cluster_seq_dict = {}
    aligned_seq_file = SeqIO.parse('seqs.aligned.fa','fasta')
    for sequence in aligned_seq_file:
        if sequence.id in cluster:
            cluster_seq_dict[sequence.id] = sequence.seq
    clusters_seq_dict.append(cluster_seq_dict)
```

```python
# count_atcg function:
# purpose: to count the number of AT, CG, or indels at each position.
# input: a dictionary of one cluster: key - sequence id, value - sequence
# output: a dictionary of AT, CG, and indel count: key - position, value - [AT count, CG count, indel count]
# output also returns the number of sequences in one cluster for later use

sequence_length = 50
def count_atcg(cluster_dict):
    ATCG_cluster = {}
    # create a dictionary with values being zeroes 
    for position in range(sequence_length):
        ATCG_cluster[position] = [0,0,0]
    # examine a piece of sequence from position 400 to position 600 to have a clearer view on the plots
    for seq in cluster_dict.values():
        seq = seq[400:450]
        for position in range(len(seq)):
            nucleotide = seq[position]
            if nucleotide == 'A'or nucleotide == 'T':
                ATCG_cluster[position][0] += 1
            elif nucleotide == 'G'or nucleotide == 'C':
                ATCG_cluster[position][1] += 1
            elif nucleotide == '-':
                ATCG_cluster[position][2] += 1            
    return ATCG_cluster,len(cluster_dict)
```

```python
# normalization function:
# purpose: normalize the results from function count_atcg and convert them to fractions
# input: results from count_atcg
# output: 3 lists of fractions of AT, CG, NA at each position 

def normalization(ATCG_cluster,num_seq_in_one_cluster):
    # create empty lists to store the fractions
    AT = [None]*sequence_length
    CG = [None]*sequence_length
    NA = [None]*sequence_length
    for position in range(sequence_length):    
        AT[position] = ATCG_cluster[position][0]/num_seq_in_one_cluster
        CG[position] = ATCG_cluster[position][1]/num_seq_in_one_cluster
        NA[position] = ATCG_cluster[position][2]/num_seq_in_one_cluster
    return AT, CG, NA
```

```python
# plot function:
# purpose: plot the fractions of AT, CG, and indel at chosen positions
# input: results from normalization function
# output: barplot stacking fractions of AT, CG, and indel at chosen positions

import numpy as np
import matplotlib.pyplot as plt

def plot(AT,CG,NA):
    positions = list(range(400,450))
    AT = np.array(AT)
    CG = np.array(CG)
    NA = np.array(NA)

    plt.bar(positions, AT, color = 'darkgreen', label = '% AT')
    plt.bar(positions, CG, bottom = AT, color = 'lightblue', label = '% CG')
    plt.bar(positions, NA, bottom = AT + CG, color = 'indianred', label = '% indel')
    plt.ylabel('%AT %GC %indel')
    plt.xlabel('position in sequence')    
    plt.legend()
    plt.show()
```

```python
# generate 12 graphs for each of the 12 cluster identified previously 

for cluster in clusters_seq_dict:
    cluster_num = clusters_seq_dict.index(cluster)
    cluster_dict = clusters_seq_dict[cluster_num]
    atcg,length = count_atcg(cluster_dict)
    AT,CG,NA = normalization(atcg,length)
    plot(AT,CG,NA)
```

```python
from Bio import AlignIO
import pandas as pd

# store the lengths of all sequences from the unaligned sequence file into a dictionary:
# key - sequence id, value - sequence length 

unaligned_seq_file = SeqIO.parse('newseqs.fa','fasta')
all_seq_length = {}
for sequence in unaligned_seq_file:
    all_seq_length[sequence.id] = len(sequence.seq)    

# group sequence lengths obtained above by cluster into a dictionary:
# key - cluster name, value - a list of sequence lengths

length_by_clusters = {}
for cluster in clusters:
    seq_length_in_one_cluster = []
    index = clusters.index(cluster)
    cluster_name = 'C' + ' ' + str(index + 1)
    for seq_name in cluster:
        seq_length = all_seq_length[seq_name]
        seq_length_in_one_cluster.append(seq_length)
    length_by_clusters[cluster_name] = seq_length_in_one_cluster

# create a pandas dataframe to examine the lengths in a table first
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in length_by_clusters.items() ]))
print(df)

df.plot.box()
plt.ylabel('Sequence Length')
plt.xlabel('Cluster') 
plt.show()
```

We can see that in most clusters, the lengths are identical. The distributions are thus uniform and are displayed as single lines instead of boxes.
