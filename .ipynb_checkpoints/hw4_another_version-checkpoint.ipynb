```python
from Bio import Entrez
from Bio import SeqIO
import timeit
import xmltodict
org = ['homo sapiens', 'escherichia coli', 'drosophila']



pwt = {'TCA cycle': ['EC:2.3.1.12','EC:1.1.1.42', 'EC:1.2.4.2', 'EC:1.1.1.37'],
       'glycolysis': ['EC:5.4.2.2', 'EC:5.1.3.3', 'EC:2.7.2.3', 'EC:4.2.1.11'],
       'pentose phosphate': ['EC:5.3.1.9', 'EC:2.2.1.1', 'EC:3.1.3.11', 'EC:4.1.2.4']}


for (path,enz) in pwt.items():
    for e in enz:
        for o in org:
           
            print('Pathway: ' + path + '; EC: ' + e + '; Organism: ' + o)

            handle = Entrez.esearch(db = 'gee',
                        term = e[3:] + ' ' + o,
                        sort = 'relevance',
                        idtype = 'acc',
                        retmax = 1)

            
            for i in Entrez.read(handle)['IdList']:
                handle = Entrez.efetch(db = 'nucleotide', id = i, rettype = 'gb', retmode = 'text')
                record = SeqIO.read(handle, "genbank")
                handle.close()
                print(str(record.seq))
                print()

              
```
