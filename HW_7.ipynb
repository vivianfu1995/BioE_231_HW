## Simulating the data
Generate binary files with different percentages of zeros and ones and protein sequence. 

````python
import numpy as np
percentages = [0.5,0.6,0.7,0.8,0.9,1]
for percentage in percentages:
    boolean = np.random.choice([0, 1], size=8*1024*1024*100, replace=True, p=[percentage, 1-percentage]) 
    boolean = np.packbits(boolean)
    name = 'zeros_'+str(int(percentage*100))+'p'
    # write this data to a file in home directory
    open(name, 'wb').write(boolean)
````

````python
import numpy as np
nt_sequence = np.random.choice(['A','T','C','G'], size=100000000, replace=True, p=[0.25,0.25,0.25,0.25])
open('nt_sequence.fa','w').write(''.join(nt_sequence))

pt_codes = ['A','R','N','D','C','E','Q','G','H','I','L','K','M','F','P','S','T','W','Y','V']
pt_sequence = np.random.choice(pt_codes, size=100000000, replace=True, p=[0.05]*20)
open('pt_sequence.fa','w').write(''.join(pt_sequence))
````

## Questions
- Which algorithm achieves the best level of compression on each file type?
    - For nucleotide sequence and protein sequence files, Arithmetic compress works the best. For files containing only zeros and ones, Arithmetic compress also achieves better level of compression compared to other methods. As the percentage of zero increases. At high percentages of zeros (50% & 40%), the advantage of using Arithmetic compression is not as obvious. 
    
    
- Which algorithm is the fastest?
    - pbzip2 is the fastest algorithms among all.
    
    
- What is the difference between bzip2 and pbzip2? Do you expect one to be faster and why?
    - pbzip2 is a parallel implementation of the bzip2 block-sorting file compressor. bzip2 compresses data in blocks and uses the Burrows–Wheeler transform to convert frequently-recurring character sequences into strings of identical letters. It then applies move-to-front transform and Huffman coding. pbzip2 is a modified version of bzip2 that supports multi-threading, giving almost linear speed improvements. In this case, pbzip2 is expected to work faster, which is confirmed by the results summarized in the first table.
    
    
- How does the level of compression change as the percentage of zeros increases? Why does this happen?
    - As the percentage of zeros increases, the level of compression also increases. This happens since fewer bits are required to encode 0.
    
    
- What is the minimum number of bits required to store a single DNA base?
    - The minimum number of bits required to store a single DNA base is $log_{2}(4) = 2 bits$, since there are four possible bases.
    
    
- What is the minimum number of bits required to store an amino acid letter?
    - The minimum number of bits required to store an amino acid letter is $log_{2}(20) = 4.32bits$, since there are 20 amino acids.
    
    
- In your tests, how many bits did gzip and bzip2 actually require to store your random DNA and protein sequences?
    - For a random DNA sequence 100 million long, the bits required should be $2 bits/nt * 100*10^6 nt = 200 * 10^6 bits = 200 MB$ in theory. Similarly, the bits required for encoding a protein sequence with the same length should be $4 bits/AA * 100*10^6 AA = 400 * 10^6 bits = 400 MB$ in theory if we approximate 4.32 bits to 4 bits per amino acid. gzip is able to achieve 29.2 MB and 60.6 MB for DNA and protein sequences respectively, while bzip2 requires 27.3 MB and 55.3 MB in reality.
    
    
- Are gzip and bzip2 performing well on DNA and proteins?
    - In turns of level of compression, gzip and bzip2 give similar performances compared to pbzip2 and Arithmetic Compress.  

## Compressing real data 
Find  the nucleic acid sequences of gp120 homologs from at least 10 different HIV isolates and concatenate them together into a single multi-FASTA.

````python
from Bio import Entrez
from Bio import SeqIO
Entrez.email = 'vivianfu@berkeley.edu'
handle = Entrez.esearch(db='nucleotide',
                        term = 'gp120',
                        sort = 'relevance',
                        idtype = 'acc'
                       )                      

record = Entrez.read(handle)
ID = record['IdList']
seqs = ''

for i in ID:
    handle = Entrez.efetch(db = 'nucleotide', id = i, rettype = 'fasta', retmode = 'text')
    record = next(SeqIO.parse(handle, 'fasta'))
    seqs += str(record.seq)

open('gp120.fa','w').write(seqs)

````

## Estimating compression of 1000 terabytes
1000 TB of data to begin with. Among which:
- 800 TB Genomes and plasmids with high similarity: 
    - Use gzip. gzip can achieves the highest compression ratio for real DNA data as seen above. Assume that the compression ratio is 77.7% (same as gp120), the reduction in data in this case is $800{TB}* 77.7\% = 621.6{TB}$.


- 100 TB Proteins: 
    - use pbzip2. pbzip2 and bzip2 work equally well for compressing protein sequences. Although ArithmeticCompress work slightly better based on the output file size results for random protein sequence, the difference is very small. Since pbzip2 works significantly faster than ArithmeticCompress, we choose pbzip2 to compress protein data assuming the compression ratio to be $(100{MB}-55.3{MB})/100{MB} = 44.7\%$. The reduction in data for protein sequence is $100{TB}*44.7\%=44.7{TB}$.


- 100 TB Random binary data: 
    - use ArithmeticCompress. Based on the results for files with random percentages of zeros, ArithmeticCompress seems to produce the best level of compression. For a completely random binary file, the compression seems to stay in the range of 0% ~ 10% based on observation from the previous results. We assume the compression to be 10% here. The reduction in data is $100{TB}*10\%=10{TB}$.
    
The overall reduction in data is $621.6+44.7+10 = 676.3{TB}$. This translates to 67.63 % reduction in data. If 500 dollars bonus is rewarded for every 1% reduction in data, the total bonus would be $(67.63\%/1\%)*{$500} = {$33,815}$.
